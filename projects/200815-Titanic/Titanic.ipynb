## 데이터마이닝 방법론 

### CRISP-DM

#### 1) 비즈니스 이해(Business understanding)
이 초기의 단계는 프로젝트 목적의 이해와 비즈니스적 관점에서 지식을 데이터 마이닝 문제 정의를 위해 전환하고 목적 달성을 위한 예비 계획을 수립하는 것을 요구한다.

#### 2) 데이터 이해(Data understanding)
데이터 이해 단계는 초기 데이터 수집을 시작으로 데이터 파악을 위한 활동, 데이터의 품질 확인, 데이터에서 통찰력 발견 및 또는 숨겨진 정보를 위한 가정으로부터 흥미 있는 서브셋 탐색으로 진행된다.

#### 3) 데이터 준비(Data preparation)
데이터 준비 단계는 초기의 데이터로부터 최종 데이터셋을 구성하기 위한 모든 활동들을 커버한다. 데이터 준비 과제에서 다중으로 업무가 수행되는 것 같으나 규정된 순서는 아니다. 과제는 모델링 툴을 위한 데이터 변환과 정제뿐만 아니라 테이블, 기록과 속성 선택 등을 포함한다.

#### 4) 모델링(Modeling)
이 단계는 다양한 모델링 기법이 선택되어 적용된다. 일반적으로 동일한 데이터 마이닝 문제 유형을 위한 몇 가지 기법들이 있다. 몇몇 기법들은 데이터 형태상의 특정 요구사항을 갖고 있다. 게다가 데이터 준비 단계로 되돌아가는 것이 종종 필요하다.

#### 5) 평가(Evaluation)
프로젝트에서 이 단계는 데이터 분석 관점으로부터 높은 품질을 갖는 모델을 만든다. 최종 전개를 진행하기 전에 만들어진 모델을 평가하고 재검토하는 철저한 평가가 꼭 필요하다.

#### 6) 전개(Deployment)
일반적으로 모델의 생성이 데이터 마이닝 프로젝트의 끝은 아니다. 데이터의 지식을 확대시키는 것이 모델의 목적이라면 획득된 지식은 잘 체계화되어야 하고 고객이 그것을 사용할 수 있도록 방법을 제시해야 한다.

### Age Category

### Age Category 설정

```python
df_train_new['AgeBand'] = pd.cut(df_train_new['Age'], 5)
df_train_new[['AgeBand', 'Survived']].groupby('AgeBand', as_index=False).mean()\
    .sort_values(by='AgeBand', ascending=True)
```

- 0: 16세 미만
- 1: 16세 이상 32세 미만
- 2: 32세 이상 48세 미만
- 3: 48세 이상 64세 미만
- 4: 64세 이상

```python
for dataset in combine_new:
    dataset.loc[dataset.Age <= 16, 'Age'] = 0
    dataset.loc[(dataset.Age > 16) & (dataset.Age <= 32), 'Age'] = 1
    dataset.loc[(dataset.Age > 32) & (dataset.Age <= 48), 'Age'] = 2
    dataset.loc[(dataset.Age > 48) & (dataset.Age <= 64), 'Age'] = 3
    dataset.loc[dataset.Age > 64, 'Age'] = 4
```

```python
df_train_new.head(6)
```

```python
df_train_new = df_train_new.drop('AgeBand', axis=1)
combine_new = [df_train_new, df_test_new]
df_train_new.head(6)
```

### Embarked(승선 항구 코드) str $\rightarrow$ int

```python
for dataset in combine_new:
    dataset['Embarked'] = dataset['Embarked']\
        .fillna(df_train_new.Embarked.dropna().mode()[0])

df_train_new[['Embarked','Survived']].groupby('Embarked', as_index=False).mean()\
    .sort_values(by='Survived', ascending=False)
```

```python
for dataset in combine_new:
    dataset['Embarked'] = dataset['Embarked']\
        .map({'S':0, 'C':1, 'Q':2}).astype(int)

df_train_new.head(6)
```

```python
df_test_new['Fare'].fillna(df_test_new['Fare'].dropna().median(), inplace=True)
df_test_new.head()
```

#### Fare 변수 Category화

```python
for dataset in combine_new:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31.0), 'Fare'] = 2
    dataset.loc[ dataset['Fare'] > 31.0, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)

df_train_new = df_train_new.drop('Fareband', axis=1)
combine_new = [df_train_new, df_test_new]

df_train_new.head(10)
```

## 4. Modeling & Assess

```python
X = df_train_new.drop(['Survived'], axis=1, inplace=False)
Y = df_train_new.Survived

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .2)
```

```python
X_train.iloc[:,0].count(), X_test.iloc[:,0].count()
```

### Support Vector Machines

```python
svc = SVC(gamma='scale')
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_train, Y_train) * 100, 2)
acc_svc
```

### KNN

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
acc_knn
```

### Logistic Regression

```python
logreg = LogisticRegression(solver='lbfgs')
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_logreg = round(logreg.score(X_train, Y_train) * 100, 2)
acc_logreg
```

### Gaussian Naive Bayes

```python
gaussian = GaussianNB()
gaussian.fit(X_train, Y_train)
Y_pred = gaussian.predict(X_test)
acc_gaussian = round(gaussian.score(X_train, Y_train) * 100, 2)
acc_gaussian
```

### Perceptron

```python
perceptron = Perceptron(max_iter=10000, tol=1)
perceptron.fit(X_train, Y_train)
Y_pred = perceptron.predict(X_test)
acc_perceptron = round(perceptron.score(X_train, Y_train) * 100, 2)
acc_perceptron
```

### Linear SVC

```python
linear_svc = LinearSVC(max_iter=10000, random_state=42)
linear_svc.fit(X_train,Y_train)
Y_pred = linear_svc.predict(X_test)
acc_linear_svc = round(linear_svc.score(X_train, Y_train) * 100, 2)
acc_linear_svc
```
